## Project Overview and Scope

One of the foundational scholarly projects of the digital humanities has been to leverage the possibilities offered by computing technologies to develop critical tools, infrastructures, and methodological pipelines that enable new forms of inquiry within the humanities. As [Piotrowski (2012)](#https://link.springer.com/chapter/10.1007/978-3-031-02146-6_2)  argues, the increasing convergence between natural language processing (NLP) and digital humanities reflects a shift on the NLP side toward genres beyond newspaper and newswire texts towards biomedical corpora, forum posts, and social media data, where computational linguists have increasingly turned their attention to historical texts and other textual forms of interest to the humanities and social sciences [(Piotrowski 2012, 7)](#(#https://link.springer.com/chapter/10.1007/978-3-031-02146-6_2).

This project engages with one such textual data, that is, the metadata produced by the social film platform Letterboxd. Originally conceived as a ‚ÄúGoodreads for film,‚Äù Letterboxd is as a social cataloguing platform  for movies, where cinephiles maintain personal film diaries, write and engage with movie reviews, and curate recommendation lists. Its large-scale collection of movie metadata, including synopses, tags, and user-generated reviews, constitutes a rich corpus for computational text analysis. From a digital humanities perspective, Letterboxd offers a valuable site for examining film genres, audience reception, fan discourses, and curatorial practices such as review-writing in contemporary digital culture.

**The scope of this project is to perform binary text classification of movie genres based on plot synopses of films released in the twenty-first century (2000‚Äì2025)**. Text classification can be defined as class of supervised machine-learning methods that assign predefined categories (two in the case of binary categories) to textual data using computational classifiers. Within both computational linguistics and digital humanities, text classification is a well-established methodological approach. Recent work by [Rich (2025)](#https://www.degruyterbrill.com/document/doi/10.1515/opth-2025-0052/html), for example, demonstrates the applicability of neural-network-based text classification in religious and medieval studies. While genre classification has been extensively explored in NLP, most existing studies focus on multi-label genre prediction or broad genre taxonomies, often using film scripts or large plot datasets. A smaller body of work has examined movie genre classification using plot summaries specifically (e.g., [Blackstock and Spitz 2008](#https://www.semanticscholar.org/paper/Classifying-Movie-Scripts-by-Genre-with-a-MEMM-Blackstock-Spitz/35355f27b62269bf8ffb260bbb5ca6a74d4ef1b9); [Abimbola 2020](#https://github.com/Wonuabimbola/movie-genre-prediction?tab=readme-ov-file); [Kumar et al. 2022](#https://link.springer.com/article/10.1007/s11042-022-13211-5)), yet these studies tend to view genre as a predictability classifier, rarely engaging with genre as a culturally constructed category that also includes subgenres like romantic comedy, horror-comedies,documentary-drama, etc.

This project focuses on romantic comedy for binary text classification using statistical and neural paradigms in machine learning. Despite its highly commercialized and conventional narrative structures and its canonical cultural significance in popular cinema, this subgenre has, to the best of my knowledge, not received focused attention as a standalone classification task. This project isolates the romantic-comedy (rom-com) genre and frames genre detection as a binary classification task‚Äîdistinguishing rom-coms from non-rom-com films using synopses of movies released from 2000 to 2025 in the Letterboxd dataset.

## Project Methodologies and Models: Naive Bayes and DistilBERT

To explore the binary text classification of romantic comedies versus non-romantic comedies, this project employs two contrasting NLP approaches: **(1) a Naive Bayes classifier using both bag-of-words and TF-IDF, and (2) DistilBERT, a transformer-based language model**.

Naive Bayes represents a classical probabilistic approach to text classification and has been widely used in foundational NLP research, including early work on movie genre classification (Blackstock and Spitz 2008). Its suitability for this project lies in both its methodological simplicity and interpretability.

Since Letterboxd synopses function as marketable metadata texts, designed to attract audiences by signalling genre conventions, Naive Bayes allows for close examination of surface-level lexical patterns using **Bag-Of-Words (BOW) and Term Frequency-Inverse Document Frequency (TF-IDF) approaches**. In the former, the model solely focuses on word frequency, while the latter can be considered an enhancement of the BOW model as it also considers the rarity of words across the dataset by not only considering word frequency but also using inverse document frequency to adjust the weight of each word before transforming it into a vector. Together, these approaches make it possible to identify which lexical features most strongly signal the rom-com genre. From a digital humanities perspective, this interpretability is particularly valuable, as it supports critical reflection on how genre is linguistically encoded within movie synopsis metadata.

By contrast, DistilBERT is a contemporary neural, transformer-based language model that captures contextual and semantic relationships between words beyond frequency-based representations. Its inclusion in the project allows for an assessment of whether deeper contextual embeddings offer advantages when classifying genre from synopsis texts.

## 2.2 Dataset Preprocessing

### 2.2.1 Modifying Columns

```python
class MovieDataPreprocessor:

    def __init__(self, df):
        self.df = df.copy()

    #remove username from reviews as an ethical measure
    def anonymize_reviews(self):
        cleaned_reviews = []

        for review_list in self.df['reviews']:
            if review_list is None:
                cleaned_reviews.append(None)
            else:
                anon_list = []
                for review in review_list:
                    anon_list.append({
                        'likes': review.get('likes', ''),
                        'review_text': review.get('review_text', '')
                    })
                cleaned_reviews.append(anon_list)

        self.df['reviews'] = cleaned_reviews
        return self

    #convert year from string to integer, if "None", make it a null value
    def convert_year(self):
        years_numeric = []

        for y in self.df['year']:
            try:
                if y is None or str(y).strip() == "None":
                    years_numeric.append(None)
                else:
                    y_str = str(y).strip()
                    if "‚Äì" in y_str:
                        y_str = y_str.split("‚Äì")[0]
                    years_numeric.append(int(y_str))
            except:
                years_numeric.append(None)

        self.df['year_numeric'] = years_numeric
        return self

    # same as above for ratings
    def convert_ratings(self):
        numeric_ratings = []

        for r in self.df['rating']:
            if r is None or str(r).strip() == "None":
                numeric_ratings.append(None)
            elif "out of" in r:
                try:
                    numeric_ratings.append(float(r.split(" out of ")[0]))
                except:
                    numeric_ratings.append(None)
            else:
                numeric_ratings.append(None)

        self.df['rating_numeric'] = numeric_ratings
        return self

    # count number of reviews per movie (incase required, optional)
    def count_reviews(self):
        review_counts = []

        for review_list in self.df['reviews']:
            if review_list is None:
                review_counts.append(0)
            else:
                review_counts.append(len(review_list))

        self.df['num_reviews'] = review_counts
        return self

    # remove duplicate titles (in case required, optional)
    def remove_duplicate_titles(self):
        self.df = self.df.drop_duplicates(subset='title', keep='first')
        return self

    # final output
    def get_clean_data(self):
        return self.df
```

## 2.3 Genre Co-Occurences in Multi-Genres Movies (2000-25)

Multi-genre movies are defined here as films that have more than one genre recorded in the dataset. The first step in analyzing genre relationships is to count co-occurrences‚Äîthat is, how often two genres are labelled together within the same movie. This provides a foundational view of subgenres in dataset.

We first get an idea of how more than one genre is listed in the dataset and then we build a co-occurrence matrix for genres in multi-genre movies (movies with ‚â•2 genres). This matrix will show how often each pair of genres appears together.

[Reference Used to Understand Co-Occurence Matrices](#https://www.geeksforgeeks.org/nlp/co-occurence-matrix-in-nlp/?utm_source=chatgpt.com)

### 2.3.1 Preparing Co-Occurences

```python
#How does genre column look like in 2000-25?
df_recent = main_df[(main_df['year_numeric'] >= 2000) & (main_df['year_numeric'] <= 2025)]
print(df_recent['genres'].head(10))
print(type(df_recent['genres'].iloc[0]))
```

```python
def build_genre_cooccurrence(df, genre_col='genres', min_genres=2):
    #df is genre column
    #genre column list/array of genres
    #min genre #minimum genre

    # collect all unique genres
    all_genres = set()
    for genres in df[genre_col]:
        if isinstance(genres, (list, np.ndarray)) and len(genres) >= min_genres: #if genre is an array and not empty
            for g in genres:
                all_genres.add(g)

    #sort genres
    all_genres = sorted(all_genres)

    #initialize matrix
    #method inspiration
    #https://www.geeksforgeeks.org/nlp/co-occurence-matrix-in-nlp/
    #we didnt have to preprocess it because it is genre labels, not corpus
    co_matrix = pd.DataFrame(
        0,
        index=all_genres,
        columns=all_genres
    )

    # fill co-occurrences
    for genres in df[genre_col]:
        if isinstance(genres, (list, np.ndarray)) and len(genres) >= min_genres:
            for i in range(len(genres)):
                for j in range(i, len(genres)):
                    g1 = genres[i]
                    g2 = genres[j]

                    co_matrix.loc[g1, g2] += 1
                    if g1 != g2:
                        co_matrix.loc[g2, g1] += 1  # keep symmetric

    return co_matrix

#find top coocurring genres (min 5) for a given genre in a multi-genre film
#so if finds comedy is coocurrencing a lot wiht drama, comedy, romance etc.
def top_cooccurring_genres(co_matrix, genre, top_n=5):
    if genre not in co_matrix.columns:
        print(f"{genre} not found in matrix.")
        return None

    result = co_matrix[genre].sort_values(ascending=False)

    # remove self-cooccurrence
    result = result.drop(genre)

    return result.head(top_n)
```

### 2.3.2 Building Co-Occurences of Top Multi-Genres in Multi-Genre Movies Released in 2000-2025

We find the top 10 genres using the co-occurence matrix function created above:

- First produce a matrix showing how often each pair of genres appears together in multi-genre movies from 2000‚Äì2025.
- Count how often each genre appears in multi-genre movies.

## 2.4 Data Subset Selection and Assigning Class Labels for Romcoms and Other Top Multi-Genres (2000-25)


For the binary classification task, a subset of movies from 2000‚Äì2025 was selected based on the **top co-occurring subgenres** during this period. The selected subgenres are:

- `'RomCom': ['Romance', 'Comedy']`
- `'Comedy-Drama': ['Comedy', 'Drama']`
- `'Thriller-Drama': ['Thriller', 'Drama']`
- `'Horror-Thriller': ['Horror', 'Thriller']`
- `'Crime-Drama': ['Crime', 'Drama']`

These subgenres were chosen because they represent the most frequent and distinctive multi-genre combinations in this era. As this is a beginner-level exploratory text classification for me, I **excluded Romantic-drama movies** to avoid overlapping with RomComs and to simplify the binary labeling scheme. This allows the model to focus on distinguishing the romantic-comedy subgenre from other prominent multi-genre categories.  

Binary labels were assigned to indicate whether a movie belongs to the RomCom class:

- `is_romcom = 1` ‚Üí movie is a romantic comedy (`'RomCom'`)
- `is_romcom = 0` ‚Üí movie belongs to any other selected subgenre  

The resulting **NLP dataframe** for model training and evaluation contains the following columns:

- `title` ‚Üí Movie title  
- `synopsis` ‚Üí Plot synopsis used as the input text  
- `genre` ‚Üí Target label representing the assigned subgenre  
- `is_romcom` ‚Üí Binary target (1 = RomCom, 0 = Other)  
- `year_numeric` ‚Üí Release year as an integer (nullable if missing)  

This structure ensures that the NLP models can efficiently process the text and learn to discriminate RomComs from other top subgenres in the dataset.


```python

#define target genres
target_genres = {
    'RomCom': ['Romance', 'Comedy'],
    'Comedy-Drama': ['Comedy', 'Drama'],
    'Thriller-Drama': ['Thriller', 'Drama'],
    'Horror-Thriller': ['Horror', 'Thriller'],
    'Crime-Drama': ['Crime', 'Drama']
}

# check keywords of genres in the genre array
def matches_genres(genre_array, keywords):

    #returns True if all keywords appear in the genre array (case-insensitive).
    #extra genres are ignored. so ['Romanctic' 'Comedy' 'Music'] is still romcom

    # make sure genre_array is a list or numpy array
    if not isinstance(genre_array, (list, np.ndarray)):
        return False

    #converts each genre to lowercase
    genre_lower = []
    for g in genre_array:
        genre_lower.append(str(g).lower())

    #checks if each keyword is present
    for kw in keywords:
        if kw.lower() not in genre_lower:
            return False

    return True

#building the df for nlp
def build_nlp_dataframe(df, target_genres):

    nlp_rows = []

    # loop over all rows in the dataframe
    for idx in range(len(df)):
        row = df.iloc[idx]

        genre_array = row['genres']
        title = row['title']
        synopsis = row.get('synopsis', '')
        year = row.get('year_numeric', np.nan)

        # loop over each target genre
        for genre_label in target_genres:
            keywords = target_genres[genre_label]

            # check if this row matches the genre
            if matches_genres(genre_array, keywords):
                # add row to the NLP list
                nlp_row = {}
                nlp_row['title'] = title
                nlp_row['synopsis'] = synopsis
                nlp_row['genre'] = genre_label

                if genre_label == 'RomCom':
                    nlp_row['is_romcom'] = 1
                else:
                    nlp_row['is_romcom'] = 0

                nlp_row['year_numeric'] = year
                nlp_rows.append(nlp_row)

    # convert list of rows into a DataFrame
    nlp_df = pd.DataFrame(nlp_rows)

    # convert year_numeric to nullable Int for readability
    nlp_df['year_numeric'] = nlp_df['year_numeric'].astype('Int64')

    return nlp_df

# build nlp df
nlp_df = build_nlp_dataframe(df_recent, target_genres)

# check df
print("NLP dataframe shape:", nlp_df.shape)


#checking genre counts
print("\nGenre counts:")

genre_counts = nlp_df['genre'].value_counts()
print(genre_counts)

#save to csv
csv_path = "nlp_df.csv"

if os.path.exists(csv_path):
    print("\nCSV already exists")
else:
    nlp_df.to_csv(csv_path, index=False)
    print("\nSaved NLP dataframe to nlp_df.csv")
```

## 2.5 Data Splitting into Train and Test for NLP Task

To ensure a **balanced dataset** for the binary classification task, we first calculated a sampling plan for non-RomCom movies. For each non-RomCom genre, we determined the total number of movies and applied a **proportional sampling fraction** to decide how many samples to retain from that genre. The resulting plan specifies a target sample size for each genre, ensuring that the number of non-RomCom instances is comparable to the number of RomCom instances. This step helps mitigate class imbalance, which can bias the model towards the majority class.

**Procedure:**
1. Count the total number of movies in each non-RomCom genre.  
2. Multiply the count by the predefined `sampling_fraction` to determine the target sample size.  
3. Round the sample size to the nearest integer and store it in a dictionary as the **sampling plan**.  
4. This plan serves as a guideline for proportional sampling in the next step.

### 2.5.1 Separating Romcom vs not Romcom

```python
romcom_df = nlp_df[nlp_df['is_romcom'] == 1]
not_romcom_df = nlp_df[nlp_df['is_romcom'] == 0]

num_romcom = len(romcom_df)

print("RomCom count:", num_romcom)
print("Not-RomCom count:", len(not_romcom_df))
```

### 2.5.2 Calculate Sampling Fraction

```python
# total number of non-romcom rows
total_not_romcom = len(not_romcom_df)

# fraction to sample from each non-romcom genre
sampling_fraction = num_romcom / total_not_romcom

print("Sampling fraction:", round(sampling_fraction, 3))
```

### 2.5.3 Sampling Plan Per Non-RomCom Genre

```python
sampling_plan = {}

for genre_name in not_romcom_df['genre'].unique():
    genre_count = len(not_romcom_df[not_romcom_df['genre'] == genre_name])

    # calculate proportional sample size
    sample_size = genre_count * sampling_fraction

    # round to a clean integer
    sample_size = int(round(sample_size))

    sampling_plan[genre_name] = sample_size

print("\nCalculated sampling plan:")
for genre_name in sampling_plan:
    print(genre_name, "‚Üí", sampling_plan[genre_name])
```

### 2.5.4 Randomized Sampling Using Sample Plan for Balanced Dataset Between Romcoms and Non-Romcoms

Once the sampling plan was established, we applied **randomized sampling** to select the required number of movies from each non-RomCom genre. This ensures that the dataset remains balanced while introducing randomness to avoid selection bias.

**Procedure:**
1. For each non-RomCom genre, retrieve the corresponding subset of movies.  
2. Compare the number of available movies to the target size in the sampling plan.  
   - If the number of available movies is greater than or equal to the target, randomly sample the exact target number.  
   - If fewer movies are available than the target, include all available movies as a safety measure.  
3. Append the sampled subsets to a combined list to form the balanced non-RomCom dataset.  
4. This randomized approach maintains representativeness across genres while supporting a fair binary classification between RomComs and non-RomComs.

```python
sampled_not_romcom_rows = []

for genre_name in sampling_plan:
    target_size = sampling_plan[genre_name]

    genre_subset = not_romcom_df[not_romcom_df['genre'] == genre_name]
    available = len(genre_subset)

    print(f"{genre_name}: available={available}, target={target_size}")

    if available >= target_size:
        sampled_subset = genre_subset.sample(
            n=target_size,
            random_state=42
        )
    else:
        sampled_subset = genre_subset  # safety measure

    sampled_not_romcom_rows.append(sampled_subset)
```

```python
balanced_df = pd.concat(
    [romcom_df] + sampled_not_romcom_rows,
    ignore_index=True
)

print("\nFinal dataset size:", balanced_df.shape)
print("\nFinal class balance:")
print(balanced_df['is_romcom'].value_counts())
```

### 3.1.2 Tokenization of Synposis Using Spacy


```python
#tokenization function
import spacy

nlp = spacy.load("en_core_web_sm")

def tokenize_texts(texts, remove_proper_nouns=False):
    """
    Tokenizes, lemmatizes, removes punctuation/stopwords from a list of texts.
    Optionally removes proper nouns (PROPN).
    Returns a list of cleaned text strings (ready for CountVectorizer).
    """
    cleaned_texts = []

    for doc in nlp.pipe(texts, disable=["ner", "parser"]):
        tokens = []
        for token in doc:
            if token.is_punct or token.is_space or token.is_stop:
                continue
            if remove_proper_nouns and token.pos_ == "PROPN":
                continue
            tokens.append(token.lemma_.lower())
        cleaned_texts.append(" ".join(tokens))

    return cleaned_texts
```

## 5.2 Dataset Class

```python
class RomcomDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=256):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_len,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].squeeze(0),  # remove batch dim
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```

## 5.3 Tokenize and Create Validation Dataset


- Using Uncased Distilbert to lowercase
- 10% validation dataset from train

```python
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# Split training set: 90% train, 10% validation
# 90% train, 10% validation split
train_idx, val_idx = train_test_split(
    bert_train.index, test_size=0.1, random_state=42, shuffle=True
)

train_texts = bert_train.loc[train_idx, 'synopsis'].tolist()
train_labels = bert_train.loc[train_idx, 'is_romcom'].tolist()

val_texts = bert_train.loc[val_idx, 'synopsis'].tolist()
val_labels = bert_train.loc[val_idx, 'is_romcom'].tolist()

test_texts = bert_test['synopsis'].tolist()
test_labels = bert_test['is_romcom'].tolist()

```

## 5.4 Dataset Objects and Loaders

```python
train_dataset = RomcomDataset(train_texts, train_labels, tokenizer)
val_dataset = RomcomDataset(val_texts, val_labels, tokenizer)
test_dataset = RomcomDataset(test_texts, test_labels, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

## 5.5 Intialize Model

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased',
    num_labels=2
)
model.to(device)

optimizer = AdamW(model.parameters(), lr=5e-5)
```

## 5.6 Train Model

```python
import torch
from sklearn.metrics import accuracy_score
from tqdm import tqdm

best_val_acc = 0.0  # keep track of best validation accuracy

epochs = 3  # or however many you want

for epoch in range(epochs):
    model.train()
    train_losses = []

    for batch in tqdm(train_loader, desc=f"Training Epoch {epoch+1}"):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        train_losses.append(loss.item())

    avg_train_loss = sum(train_losses)/len(train_losses)

    # Validation
    model.eval()
    val_preds, val_labels = [], []
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask)
            preds = torch.argmax(outputs.logits, dim=1)

            val_preds.extend(preds.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())

    val_acc = accuracy_score(val_labels, val_preds)
    print(f"Epoch {epoch+1}: Train Loss={avg_train_loss:.4f}, Val Accuracy={val_acc:.4f}")

    # Save checkpoint if this epoch has the best validation accuracy
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        checkpoint_path = f"best_distilbert_epoch{epoch+1}.pth"
        torch.save(model.state_dict(), checkpoint_path)
        print(f"**Checkpoint saved: {checkpoint_path} (Val Accuracy={val_acc:.4f})**")
```

#### Interpretation of DistilBERT Training Model

During the fine-tuning of DistilBERT for the Romcom classification task, the model was trained for three epochs. In the first epoch, the model achieved a training loss of 0.5121 and a validation accuracy of 0.7590, which was recorded as the best checkpoint at that point. In the second epoch, the training loss decreased to 0.4120 and the validation accuracy slightly improved to 0.7596, marking the **highest validation performance** across all epochs. In the third epoch, although the training loss continued to decrease to 0.2959, the validation accuracy dropped slightly to 0.7485.

This pattern indicates that the model was increasingly fitting the training data while beginning to **overfit slightly**, as evidenced by the small decrease in validation performance. Accordingly, the best checkpoint to use for evaluation on the test set is **best_distilbert_epoch2.pth**, corresponding to the highest validation accuracy of **75.96%**. Using this checkpoint ensures that the model applied to the test set generalizes best to unseen synopses rather than the final epoch, which shows early signs of overfitting.
